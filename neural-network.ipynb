From the correlation plot of the mort_acc and other columns in the table we can see that total_acc has the highest correlation. As such, we can use values from the total_acc to fill in the null values of the mort_acc. 

To fill the null values, I will take the mean of the total value, this is because the number of total account any owner has, I will assume some of them maybe mortage accounts.

## TASK 3: Train Test Split

```python
from sklearn.model_selection import train_test_split
```

#### TASK 3.1: Set X and y variables to the values of the features and label.

```python
# need to remove the string value of month from issue_d in order to apply fit_transform
# df['issue_d'] = df['issue_d'].apply(lambda int(x:x.split('-')[-1]))
df.drop('issue_d', axis=1, inplace=True)
df.drop('loan_status', axis=1, inplace=True)
```

```python
X = df.drop('loan_repaid', axis=1).values
```

```python
y = df['loan_repaid'].values
```

## TASK 4: Grabbing a Sample for Training Time

#### TASK 4.1 : Perform a train/test split with test_size=0.2 and a random_state of 101.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

#### TASK 4.2: Use a MinMaxScaler to normalize the feature data X_train and X_test. 

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## TASK 5: Creating the Model

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

#### TASK 5.1: Build a sequential model which will be trained on the data. OPTIONAL: You can add Dropout layers in order to avoid the overfitting problem.

```python
model = Sequential([
    Dense(78,  activation='relu', name="InputLayer"),
    Dropout(0.2),
    Dense(58, activation='relu', name="HiddenLayer1"),
    Dropout(0.2),
    Dense(38, activation='relu', name="HiddenLayer2"),
    Dropout(0.2),
    Dense(18, activation='relu', name="HiddenLayer3"),
    Dropout(0.2),
    Dense(units=1,activation='sigmoid', name="OutputLayer"),
])

model.compile(loss='binary_crossentropy', optimizer='adam')
```

#### TASK 5.2 : Fit the model to the training data for at least 25 epochs and batch_size of 256. 

```python
model.fit(x=X_train, y=y_train, epochs=25, batch_size=256, validation_data=(X_test, y_test))
```

#### TASK 5.3: Save your model.

```python
model.save('loan_data_model.h5')  
```

## TASK 6: Evaluating Model Performance

#### TASK 6.1: Plot out the validation loss versus the training loss.

```python
loss = pd.DataFrame(model.history.history)
```

```python
loss[['loss','val_loss']].plot()
```

#### TASK 6.2: Create predictions from the X_test set and display a classification report and confusion matrix for the X_test set.

```python
predict_x= model.predict(X_test)
predict_x= (predict_x > .5).astype(int)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test, predict_x))
```

```python
print(confusion_matrix(y_test, predict_x))
```
